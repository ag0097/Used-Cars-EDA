#**Feature Engineering and Data Cleaning**

```python
data_df.corr()
#For YEAR, SELLING PRICE, KM DRIVEN
```

```python
print('Costliest Car was: Rs ',data_df['SELLING PRICE'].max())
print('Cheapest Car was: Rs ',data_df['SELLING PRICE'].min())
print('Average price of Car was: Rs ',data_df['SELLING PRICE'].mean())
```

**Discretization & Binning**

```python
data_df['Price_band'] = 0
data_df.loc[data_df['SELLING PRICE'] <= 150000,'Price_band'] = 0
data_df.loc[(data_df['SELLING PRICE'] > 150000) & (data_df['SELLING PRICE'] <= 300000),'Price_band'] = 1
data_df.loc[(data_df['SELLING PRICE'] > 300000) & (data_df['SELLING PRICE'] <= 500000),'Price_band'] = 2
data_df.loc[(data_df['SELLING PRICE'] > 500000) & (data_df['SELLING PRICE'] <= 800000),'Price_band'] = 3
data_df.loc[data_df['SELLING PRICE'] > 800000,'Price_band'] = 4
data_df.head(10)
```

```python
data_df.corr()
```

```python
data_df['Price_band'].value_counts().to_frame().style.background_gradient(cmap='winter')
```

```python
print('Maximum Km driven were: ',data_df['KM DRIVEN'].max(),'kms')
print('Minimum Km driven were: ',data_df['KM DRIVEN'].min(),'kms')
print('Average Km driven were: ',data_df['KM DRIVEN'].mean(),'kms')
```

```python
data_df['Km_band'] = 0
data_df.loc[data_df['KM DRIVEN'] <= 50000,'Km_band'] = 0
data_df.loc[(data_df['KM DRIVEN'] > 50000) & (data_df['KM DRIVEN'] <= 100000),'Km_band'] = 1
data_df.loc[(data_df['KM DRIVEN'] > 100000) & (data_df['KM DRIVEN'] <= 150000),'Km_band'] = 2
data_df.loc[(data_df['KM DRIVEN'] > 150000) & (data_df['KM DRIVEN'] <= 200000),'Km_band'] = 3
data_df.loc[data_df['KM DRIVEN'] > 200000,'Km_band'] = 4
data_df.head(10)
```
